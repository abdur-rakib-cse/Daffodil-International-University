```python
def Func(lines):
  lines = lines.lower()
  lines = lines.split()
  return lines

split_RDD = text_RDD.map(Func)
```

```python
split_RDD.count()
```

```python
split_RDD.take(1)
```

```python
word_split_RDD = text_RDD.flatMap(Func)
```

```python
word_split_RDD.count()
```

```python
word_split_RDD.take(5)
```

```python
stopword = ['is','to','for','a','an','the','about']
```

```python
filtered_RDD = word_split_RDD.filter(lambda x: x not in stopword)
```

```python
filtered_RDD.count()
```

```python
specific_words_RDD = word_split_RDD.filter(lambda x: x.startswith('l'))
```

```python
specific_words_RDD.count()
```

```python
specific_words_RDD.take(5)
```

```python
specific_words_RDD.distinct().count()
```

```python
specific_words_RDD.distinct().collect()
```
