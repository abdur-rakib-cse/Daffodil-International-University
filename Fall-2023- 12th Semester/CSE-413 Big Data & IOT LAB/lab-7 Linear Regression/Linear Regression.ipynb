```python
from pyspark.ml.feature import StringIndexer
```

```python
indexer = StringIndexer(inputCol = "sex", outputCol = "sex_cat")
```

```python
indexed = indexer.fit(df).transform(df)
```

```python
indexed.show()
```

```python
indexer = StringIndexer(inputCol = "smoker", outputCol = "smoker_cat")
indexed = indexer.fit(indexed).transform(indexed)
```

```python
indexer = StringIndexer(inputCol = "region", outputCol = "region_cat")
indexed = indexer.fit(indexed).transform(indexed)
```

```python
indexed.show()
```

```python
from pyspark.ml.linalg import Vector
from pyspark.ml.feature import VectorAssembler
```

```python
column_list
```

```python
assembler = VectorAssembler(inputCols=['age', 'bmi', 'children', 'sex_cat', 'smoker_cat', 'region_cat'], outputCol = "features")
```

```python
assembler
```

```python
output = assembler.transform(indexed)
```

```python
output.show(truncate = False)
```

```python
output.select("features","expenses").show(truncate = False)
```

```python
final_data = output.select("features","expenses")
```

```python
final_data.show()
```

```python
train_data, test_data = final_data.randomSplit([0.7,0.3])
```

```python
train_data.show()
```

```python
test_data.show()
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
lr = LinearRegression(featuresCol = "features", labelCol = "expenses")
```

```python
trained_model = lr.fit(train_data)
```

```python
trained_model
```

```python
results = trained_model.evaluate(train_data)
```

```python
print(results.r2)
```

```python
print(results.meanSquaredError)
```

```python
predictions = trained_model.transform(test_data)
```

```python
predictions.show()
```
