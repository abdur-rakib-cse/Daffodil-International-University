```python
import pandas as pd
import numpy as np
from pyspark.sql import SQLContext
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
cases = spark.read.load("/content/drive/MyDrive/Big Data and IoT Files/Lab Contents/Lab 4: DataFrames/Copy of Case.csv", format = "csv", sep = ",",inferScheme = True, header = True)
```

```python
cases.show()
```

```python
cases.limit(10).toPandas()
```

```python
cases = cases.withColumnRenamed("infection_case", "infection_source")
```

```python
cases.show()
```

```python
cases = cases.toDF(*['id', 'province', 'city', 'group', 'infection_case', 'confirmed_number', 'latitude', 'longitude'])
```

```python
cases.show()
```

```python
cases = cases.withColumnRenamed("confirmed_number", "confirmed")
```

```python
cases.show()
```

```python
cases = cases.select('province', 'city', 'infection_case', 'confirmed')
```

```python
cases.show()
```

```python
cases.sort("confirmed").show()
```

```python
from pyspark.sql import functions as F
cases.sort(F.desc("confirmed")).show()
```

```python
from pyspark.sql.types import IntegerType, StringType, FloatType
```

```python
cases = cases.withColumn('confirmed',F.col("confirmed").cast(IntegerType()))
```

```python
cases.show()
```

```python
cases.sort(F.desc("confirmed")).show()
```

```python
cases.filter((cases.confirmed >68) & (cases.province == 'Seoul')).show()
```

```python
cases.groupBy(["province", "city"]).agg(F.sum("confirmed").alias("total confirmed"), F.max("confirmed").alias("max confirmed cases")).show()
```

```python
regions = spark.read.load("/content/drive/MyDrive/Big Data and IoT Files/Lab Contents/Lab 4: DataFrames/Copy of Region.csv", format = "csv", sep = ",",inferScheme = True, header = True)
```

```python
regions.show()
```

```python
regions.limit(10).toPandas()
```

```python
cases.show()
```

```python
cases = cases.join(regions, ["province", "city"], how = "left")
```

```python
cases.limit(10).toPandas()
```

```python
NewConfirmed = cases.withColumn("NewConfirmed", 100 + F.col("confirmed"))
```
